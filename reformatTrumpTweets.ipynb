```python
columns = list(df)
columns
```

```python
df = df.rename(index=str, columns={columns[0]: 'temp'})
```

```python
columns = columns[0].split(',')
columns
```

```python
new_df = pd.DataFrame(columns=['id'] + columns)
keep_retweets = False
num_retweets_dumped = 0
num_tweets_and_retweets = 0
i = -1
for row in df.itertuples(index=True, name='Pandas'):
#     print()
#     print(row)
    
    i += 1  # new_df row index
    num_tweets_and_retweets += 1
    
    # get and split the content - check to make sure split is correct - exit if not
    content = getattr(row, 'temp')
    content_list = content.split(',')
    temp_length = len(content_list)
    if temp_length != 7:
        print(len(content.split(',')))
        sys.exit('wrong number of list elements - check for unexpected commas')
        
    # load the split content into new data frame
    j = -1
    if keep_retweets == False and content_list[5] == 'true':
#         print('is_retweet:', content_list[5], ';keep_retweets:', keep_retweets, 'skip tweet')
        num_retweets_dumped += 1
        continue
    else:
        new_df.at[i, 'id'] = row[0]
        for column in columns:
            j += 1
            new_df.at[i, column] = content_list[j]

print('num_tweets_and_retweets:', num_tweets_and_retweets)
print('num_retweets_dumped:', num_retweets_dumped)
```
